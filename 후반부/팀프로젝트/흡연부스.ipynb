```python
smoke['부스주소'] = smoke['명칭'] + ' ' + smoke['주소']
smoke['부스주소'].head()
```

```python
sname = []

for name in smoke['부스주소']:
    sname.append(name)
    
sname
```

```python
addreess = []
lat = []
lng = []

for name in sname:
    try :
        tmp = gmaps.geocode(name, language='ko')
        addreess.append(tmp[0].get("formatted_address"))

        tmp_loc = tmp[0].get("geometry")

        lat.append(tmp_loc['location']['lat'])
        lng.append(tmp_loc['location']['lng'])

        print(name + '-->' + tmp[0].get("formatted_address"))
    except IndexError:
        pass
```

```python
len(addreess)
```

```python
data = {'주소':addreess, '위도':lat, '경도':lng }
saddress = pd.DataFrame(data)
saddress.head()
```

```python
boothnum = pd.read_csv('teamproject/booth_num.csv', thousands=',', encoding='cp949')
boothnum
```

```python
boothnum.set_index('구', inplace = True)
boothnum.head()
```

```python
#02skorea_municipalities_geo_simple
import json
geo_path = 'data/map/skorea-municipalities-2018-geo.json'
geo_json = json.load(open(geo_path, encoding = 'utf-8'))
```

```python
map = folium.Map(location=[booth['위도'].mean(), booth['경도'].mean()], zoom_start=12, tiles='Mapbox Control Room')

map.choropleth(geo_data = geo_json,
               data = boothnum['부스수'],
               columns = [boothnum.index, boothnum['부스수']],
               fill_color = 'PuBu',
               key_on = 'feature.properties.name',
               fill_opacity=1,
               line_opacity=1,
               highlight=True)

for n in booth.index:    
    folium.CircleMarker(
        location=[booth['위도'][n], booth['경도'][n]],
        radius=10,
        color= 'blue',
        fill=True,
        fill_color='blue'
    ).add_to(map)


map.save('data/map/booth.html')
```
