# 그래프 크기 설정
- plt.figure(figsize = (가로,세로))

```python
x = np.arange(-10,10)
y = np.tan(x)
plt.figure(figsize=(10,5))
plt.plot(x, y, label='tan',linestyle=':')

plt.legend()

plt.grid()
plt.show()
```

## 마커 표시
- marker = 마커모양
- markerfacecolor = 칼라 값
- markersize = 크기

```python
x = np.arange(-10,10)
y = x**2
plt.figure(figsize=(6,4))
plt.plot(x,y,c='g',markerfacecolor = 'red', marker='D', markersize = 7)
plt.show()
```

```python
names = ['group_a','group_b','group_c']
values1 = [1, 10 ,100]
values2 = [100, 50, 10]
values3 = [100, 40, 100]

plt.figure(figsize=(9,3))
plt.subplot(131)
plt.plot(names, values1)
plt.subplot(132)
plt.plot(names, values2)
plt.subplot(133)
plt.plot(names, values3)

plt.suptitle('1*3 subplot')
plt.show()
```

```python
names = ['group_a','group_b','group_c']
values = [1, 10 ,100]

plt.figure(1, figsize=(9, 3))
plt.subplot(131)
plt.bar(names, values)
plt.subplot(132, facecolor= 'b')
plt.scatter(names, values, c='w')
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```

```python
names = ['group_a','group_b','group_c']
values1 = [1, 10 ,100]
values2 = [100, 50, 10]
values3 = [100, 40, 100]
values4 = [10, 100, 50]

fig = plt.figure(figsize=(9,3))
fig, ax = plt.subplots(2,2)
ax[0, 0].plot(names, values1)
ax[0, 1].plot(names, values2)
ax[1, 0].plot(names, values3)
ax[1, 1].plot(names, values4)
plt.suptitle('Multi Subplots')
plt.show()
```

```python
x = np.arange(-10,10)
fig1 = plt.figure(figsize=(9, 3))
fig1, plt.figure()
fig1, bx = plt.subplots(3,1)
bx[0].plot(x, np.sin(x))
bx[1].plot(x, np.cos(x))
bx[2].plot(x, np.tan(x))
plt.show()
```

```python
s1 = pd.Series([20, 35, 67, 33, 10, 56, 23])
s2 = pd.Series(np.random.randint(40,80,size=10))

fig11 = plt.figure()
fig11, ax2 = plt.subplots(1,2)
ax2[0].plot(s1)
ax2[1].plot(s2)
plt.suptitle('Series subplots')
plt.show()
```

```python
people = ('a','b','c','d','e','f')
y_pos = np.arange(len(people))
performance = np.random.rand(len(people))
fig16 = plt.figure(figsize=(4,4))
fig16, ax = plt.subplots()
ax.barh(y_pos,performance,color='g')
#눈금 값
ax.set_yticks(y_pos)
ax.set_yticklabels(people)
ax.set_xlabel('Performance')
fig16, plt.show()
```
