# Scatter Plot
- 산포도

````python
names = ['a','b','c','d','e']
values = np.random.randint(30,50,5)

plt.scatter(names, values, s=200, c='b', marker = 'X')
plt.suptitle('산포도')
plt.show()
````

````python
names = ['a','b','c','d','e']
values = np.random.randint(30,50,5)

num = len(names)

fig_s2 = plt.figure()
fig_s2, plt.scatter(names, values, color = 'r')

for i in np.arange(num):
    fig_s2, plt.text(names[i],values[i],values[i])
    
fig_s2, plt.suptitle('Scatter Plot Sample')
fig_s2, plt.show()
````

# Scatter plot

# Pie Plot
- figure 생성
- plt.pie(values, labels=labelList, autopct = '%digit.pointNum%Unit')

````python
labelList = 'a','b','c','d'
sizes = [15,30,45,10]
fig_p1 = plt.figure()
fig_p1, plt.pie(sizes, labels=labelList, autopct='%.1f%%')
plt.show()
````

# Pie Plot - Explode


````python
labelList = 'a','b','c','d'
sizes = [15,30,45,10]

explode = (0.05,0.1,0.15,0.05)

fig_p1 = plt.figure()
fig_p1, plt.pie(sizes,explode=explode, labels=labelList, autopct='%.1f%%',shadow=True,startangle=180)
plt.show()
````

````python
labelList = 'a','b','c','d'
sizes = [15,30,45,10]

colorList=['b','g','r','c']
explode = (0.05,0.1,0.15,0.05)

fig_p1 = plt.figure()
fig_p1, plt.pie(sizes,explode=explode, labels=labelList, autopct='%.1f%%',shadow=True,startangle=180,
                textprops={'fontsize': 14}, colors=colorList)
plt.show()
````

# histogram + subplot
- fig_h2, axs = plt.subplots(1, 2, sharey=True)
```
sharey=True : 간격 자동지정
```

````python
n_bins = 20

x= np.random.randn(100000)
y= np.random.randn(100000)

fig_h2 = plt.figure(figsize = (12,12))
fig_h2, axs = plt.subplots(1, 2, sharey=True)

fig_h2, axs[0].hist(x, bins=n_bins, color='yellowgreen')
fig_h2, axs[1].hist(y, bins=n_bins, color='brown')
````
